# IEP-2001.2 — 하이브리드 검색 방법 F (threshold 제거)

> **목표**: threshold 필터를 완전히 제거하고 RAGAS 측정 → 방법 E의 Recall upper bound 확인
> **방법 F**: 벡터 + BM25 → RRF → TOP_K=3 반환. 거절 없음 (threshold 필터 없음)
> **비교 대상**:
> - IEP-1001 Solar v2: Recall 0.6549 / Precision 0.5608 (Solar judge, k=10, 거절 없음)
> - IEP-2001.1 방법E: Recall 0.5600 / Precision 0.6835 (Solar judge, threshold=0.20, 거절 10/100)

## 방법 E vs 방법 F

```
[방법 E — threshold=0.20]          [방법 F — threshold 제거]
벡터 k=10 + BM25 k=10               벡터 k=10 + BM25 k=10
→ RRF → TOP_K=3                     → RRF → TOP_K=3
→ TOP_K 최고 score < 0.20 → 거절    → 거절 없음, 항상 TOP_K 반환
→ 거절 10/100                       → 거절 0/100 (upper bound)
```

**측정 목적**: 방법 E와 F의 Recall 차이가 작으면 threshold=0.20이 적절,
크면 threshold를 더 낮추거나 제거를 고려해야 함.

## Cell 5 — 검색 함수 정의

### 방법 F: threshold 제거 — 거절 없음 (IEP-2001.2 핵심)

```
[방법 E (IEP-2001.1)]                [방법 F (IEP-2001.2)]
벡터 k=10 + BM25 k=10                벡터 k=10 + BM25 k=10
→ RRF → TOP_K=3                      → RRF → TOP_K=3
→ TOP_K 최고 score < 0.20 → 거절     → 항상 TOP_K 반환 (거절 없음)
→ 거절 10/100                        → 거절 0/100
```

**변경 포인트**: `hybrid_search()`에서 threshold 체크 블록 제거


````python
import os
from langchain_core.documents import Document

# ── 파라미터 ────────────────────────────────────────────────────────
TOP_K            = 3    # 최종 반환 청크 수
BM25_CANDIDATE_K = 10   # BM25/벡터 각각에서 뽑는 후보 수
RRF_K            = 60   # RRF 하이퍼파라미터 (표준값)
# SIMILARITY_THRESHOLD 없음 — 방법 F는 threshold 제거
# ────────────────────────────────────────────────────────────────────


def vector_search(question: str, k: int = BM25_CANDIDATE_K):
    """벡터 검색 → List[(Document, cosine_score)]"""
    return vectorstore.similarity_search_with_relevance_scores(question, k=k)


def bm25_search(question: str, k: int = BM25_CANDIDATE_K):
    """BM25 검색 → List[(Document, bm25_score)]"""
    tokens = tokenize(question)
    scores = bm25.get_scores(tokens)
    top_indices = scores.argsort()[::-1][:k]
    results = []
    for idx in top_indices:
        doc = Document(
            page_content=docs_text[idx],
            metadata=docs_meta[idx]
        )
        results.append((doc, float(scores[idx])))
    return results


def rrf_fusion(vector_results, bm25_results, top_k: int = TOP_K):
    """
    Reciprocal Rank Fusion
    rrf_score = Σ 1 / (RRF_K + rank)  (rank: 0-indexed)
    """
    rrf_scores = {}
    doc_map    = {}

    def doc_key(doc: Document) -> str:
        return doc.page_content[:80]

    for rank, (doc, _) in enumerate(vector_results):
        k_ = doc_key(doc)
        rrf_scores[k_] = rrf_scores.get(k_, 0.0) + 1.0 / (RRF_K + rank + 1)
        doc_map[k_] = doc

    for rank, (doc, _) in enumerate(bm25_results):
        k_ = doc_key(doc)
        rrf_scores[k_] = rrf_scores.get(k_, 0.0) + 1.0 / (RRF_K + rank + 1)
        if k_ not in doc_map:
            doc_map[k_] = doc

    sorted_keys = sorted(rrf_scores, key=rrf_scores.get, reverse=True)
    top_keys    = sorted_keys[:top_k]
    return [(doc_map[k_], rrf_scores[k_]) for k_ in top_keys]


def hybrid_search(question: str):
    """
    하이브리드 검색 메인 함수 (방법 F: threshold 제거 — 거절 없음)

    방법 E와의 차이:
      방법 E: TOP_K 최고 벡터 score < 0.20 → 거절
      방법 F: threshold 체크 없음 → 항상 TOP_K 반환

    반환: List[(Document, rrf_score)]
    """
    v_results = vector_search(question, k=BM25_CANDIDATE_K)
    b_results = bm25_search(question,  k=BM25_CANDIDATE_K)
    return rrf_fusion(v_results, b_results, top_k=TOP_K)


print("검색 함수 정의 완료 (IEP-2001.2 방법 F — threshold 없음)")
print(f"  TOP_K={TOP_K}, CANDIDATE_K={BM25_CANDIDATE_K}, RRF_K={RRF_K}")
print(f"  거절 기준: 없음 (항상 TOP_K 반환)")
````

## Cell 7 — 골든 데이터셋 로드 + 100개 하이브리드 검색 실행

````python
import csv

with open(GOLDEN_PATH, "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    golden = list(reader)

print(f"골든 데이터셋: {len(golden)}개")
print(f"키 확인: {list(golden[0].keys())}")
print(f"샘플 question: {golden[0]['user_input'][:60]}")
print(f"샘플 reference: {golden[0]['reference'][:60]}")
````

````python
import time

print("100개 하이브리드 검색 실행 중...")
t0 = time.time()

hybrid_dataset = []
rejected_count = 0

for i, item in enumerate(golden):
    q            = item["user_input"]       # ← question → user_input
    ground_truth = item["reference"]        # ← ground_truth → reference

    results  = hybrid_search(q)
    contexts = [doc.page_content for doc, _ in results]

    if not contexts:
        rejected_count += 1

    hybrid_dataset.append({
        "question":        q,
        "ground_truth":    ground_truth,
        "contexts":        contexts,
        "retrieved_count": len(contexts),
        "rejected":        len(contexts) == 0,
    })

    if (i + 1) % 20 == 0:
        elapsed = time.time() - t0

elapsed = time.time() - t0
avg_chunks = sum(r["retrieved_count"] for r in hybrid_dataset) / len(hybrid_dataset)
print(f"거절 수: {rejected_count}/100")
print(f"평균 검색 청크 수: {avg_chunks:.2f}")
````

## Cell 8 — Solar 답변 생성 (100개)

기존 `IEP1001_case3_ragas_solar.ipynb`와 동일한 프롬프트 사용

````python
from openai import OpenAI
import time

SOLAR_BASE_URL = os.environ.get("SOLAR_BASE_URL", "https://api.upstage.ai/v1")
SOLAR_MODEL    = "solar-pro3"

solar_client = OpenAI(
    api_key=UPSTAGE_API_KEY,
    base_url=SOLAR_BASE_URL
)

RAG_PROMPT = """당신은 IPCC AR6 기후변화 보고서 전문 안내 AI입니다.
반드시 아래 제공된 문서 내용만을 근거로 답변하세요.
문서에 없는 내용은 추측하지 말고 "해당 내용은 IPCC 보고서에서 찾을 수 없습니다."라고 답하세요.

[참고 문서]
{context}

[질문]
{question}

[답변]"""

REJECTION_MESSAGE = (
    "죄송합니다. 해당 질문은 IPCC AR6 보고서의 범위를 벗어납니다. "
    "IPCC 기후변화 관련 질문을 입력해 주세요."
)

def generate_answer(question: str, contexts: list) -> str:
    if not contexts:
        return REJECTION_MESSAGE
    context_text = "\n\n".join(
        f"[청크 {i+1}]\n{c}" for i, c in enumerate(contexts)
    )
    prompt = RAG_PROMPT.format(context=context_text, question=question)
    resp = solar_client.chat.completions.create(
        model=SOLAR_MODEL,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=800
    )
    return resp.choices[0].message.content


print("Solar 답변 생성 중 (100개)...")
t0 = time.time()

for i, item in enumerate(hybrid_dataset):
    item["answer"] = generate_answer(item["question"], item["contexts"])
    if (i + 1) % 10 == 0:
        elapsed = time.time() - t0

total = time.time() - t0
print(f"\n답변 생성 완료: {total:.0f}초")
````

````python
import numpy as np
from datetime import datetime

ALL_METRICS = ["context_recall", "context_precision", "faithfulness", "answer_relevancy"]
TYPES       = ["사실 확인", "비교", "의견/예측", "범위 밖"]

# ── 기존 실험 수치 ───────────────────────────────────────────────
IEP1001_V2 = {
    "전체":     {"context_recall": 0.6549, "context_precision": 0.5608, "faithfulness": 0.3165, "answer_relevancy": 0.4075},
    "사실 확인": {"context_recall": 0.8000, "context_precision": 0.6006, "faithfulness": 0.5646, "answer_relevancy": 0.6191},
    "비교":     {"context_recall": 0.4600, "context_precision": 0.6220, "faithfulness": 0.1429, "answer_relevancy": 0.4089},
    "의견/예측": {"context_recall": 0.7014, "context_precision": 0.7623, "faithfulness": 0.3250, "answer_relevancy": 0.4867},
    "범위 밖":  {"context_recall": 0.6600, "context_precision": 0.2598, "faithfulness": 0.1520, "answer_relevancy": 0.1155},
}
IEP2001_E = {
    "전체":     {"context_recall": 0.5600, "context_precision": 0.6835, "faithfulness": 0.2597, "answer_relevancy": 0.3537, "rejected": 10},
    "사실 확인": {"context_recall": 0.7200, "context_precision": 0.8229, "faithfulness": 0.4774, "answer_relevancy": 0.5743},
    "비교":     {"context_recall": 0.3600, "context_precision": 0.7933, "faithfulness": 0.3308, "answer_relevancy": 0.3575},
    "의견/예측": {"context_recall": 0.5600, "context_precision": 0.8767, "faithfulness": 0.3699, "answer_relevancy": 0.4637},
    "범위 밖":  {"context_recall": 0.6000, "context_precision": 0.2467, "faithfulness": 0.0000, "answer_relevancy": 0.0192},
}

# ── df_raw 구성 ──────────────────────────────────────────────────
df_raw = pd.DataFrame({
    "id":                [r["id"]         for r in eval_records],
    "type":              [r["type"]        for r in eval_records],
    "user_input":        [r["user_input"]  for r in eval_records],
    "context_recall":    df_retrieval["context_recall"].values,
    "context_precision": df_retrieval["context_precision"].values,
    "faithfulness":      df_faith["faithfulness"].values,
    "answer_relevancy":  df_ar["answer_relevancy"].values,
    "rejected":          [len(r["contexts"]) == 0 for r in eval_records],
})

overall_mean  = df_raw[ALL_METRICS].mean(skipna=True).round(4)
overall_nan   = df_raw[ALL_METRICS].isna().sum()
summary       = df_raw.groupby("type")[ALL_METRICS].mean().round(4)
rejected_count = df_raw["rejected"].sum()

faith_conservative = round(
    overall_mean["faithfulness"] * (100 - overall_nan["faithfulness"]) / 100, 4
)

# ── 3-way 비교표 ─────────────────────────────────────────────────
print("=" * 72)
print("  전 지표 비교 (Solar judge 기준)")
print("=" * 72)
header = f"  {'단계':<26} {'Recall':>8} {'Precision':>10} {'Faith':>8} {'Faith(보)':>10} {'AR':>8} {'거절':>7}"
print(header)
print("  " + "-" * 68)

rows = [
    ("IEP-1001 Solar v2 (벡터)",  IEP1001_V2["전체"],  "-"),
    ("IEP-2001.1 방법E (thr=0.20)", IEP2001_E["전체"],  f"{IEP2001_E['전체']['rejected']}/100"),
    ("IEP-2001.2 방법F (thr 없음)", {m: overall_mean[m] for m in ALL_METRICS}, f"{rejected_count}/100"),
]
for label, vals, rej in rows:
    fa_con = round(vals["faithfulness"] * (100 - 0) / 100, 4) if label != "IEP-2001.2 방법F (thr 없음)" else faith_conservative
    print(f"  {label:<26} {vals['context_recall']:>8.4f} {vals['context_precision']:>10.4f} "
          f"{vals['faithfulness']:>8.4f} {fa_con:>10.4f} {vals['answer_relevancy']:>8.4f} {rej:>7}")

print()

# ── 방법E vs F 차이 ──────────────────────────────────────────────
print("  [방법E → 방법F 변화]")
for m in ALL_METRICS:
    e_val = IEP2001_E["전체"][m]
    f_val = overall_mean[m]
    arrow = "↑" if f_val > e_val else "↓"
    print(f"    {m:<22}: {e_val:.4f} → {f_val:.4f}  ({arrow}{abs(f_val-e_val):.4f})")
print(f"    {'거절':<22}: {IEP2001_E['전체']['rejected']}/100 → {rejected_count}/100")

# ── 생존 편향 보정 ───────────────────────────────────────────────
print()
print("  [생존 편향 보정]")
print(f"  {'지표':<22} {'낙관적':>10} {'유효샘플':>8} {'보수적':>10}")
print("  " + "-" * 54)
for m in ALL_METRICS:
    opt = overall_mean[m]
    vn  = 100 - overall_nan[m]
    pes = round(opt * vn / 100, 4)
    print(f"  {m:<22} {opt:>10.4f} {vn:>8}개 {pes:>10.4f}")

# ── README 마크다운 ──────────────────────────────────────────────
print()
print("[README 마크다운]")
md  = "| 유형 | Context Recall | Context Precision | Faithfulness | Answer Relevancy | 거절 |\n"
md += "| :--- | :---: | :---: | :---: | :---: | :---: |\n"
for t in TYPES:
    cr = summary.loc[t, "context_recall"]    if t in summary.index else float("nan")
    cp = summary.loc[t, "context_precision"] if t in summary.index else float("nan")
    fa = summary.loc[t, "faithfulness"]      if t in summary.index else float("nan")
    ar = summary.loc[t, "answer_relevancy"]  if t in summary.index else float("nan")
    md += f"| {t} | {cr:.4f} | {cp:.4f} | {fa:.4f} | {ar:.4f} | - |\n"
md += (f"| **전체 (방법F)** | **{overall_mean['context_recall']:.4f}** "
       f"| **{overall_mean['context_precision']:.4f}** "
       f"| **{overall_mean['faithfulness']:.4f}** "
       f"| **{overall_mean['answer_relevancy']:.4f}** | **0/100** |\n")
md += (f"| *방법E (thr=0.20)* | *0.5600* | *0.6835* | *0.2597* | *0.3537* | *10/100* |\n")
md += (f"| *IEP-1001 Solar v2* | *0.6549* | *0.5608* | *0.3165* | *0.4075* | *-* |\n")
print(md)
````
