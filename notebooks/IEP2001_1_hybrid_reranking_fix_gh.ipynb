## IEP-2001 실패 원인 분석

```
[IEP-2001 방법 C — 문제]
벡터 검색 (k=10) → 10개 결과 전체에 threshold 0.28 적용
→ 1개라도 통과 시 진행, 전체 미달 시 거절

문제: CANDIDATE_K=10 기준이므로 8~9번째 결과(score 낮음)가
     간신히 threshold 통과하는 "경계" 질문도 살아남음
     → 반대로 borderline 질문(score 0.26~0.28)은 10개 중
       최고 score도 0.28 미달 → 거절 17/100 발생

[IEP-2001.1 방법 D — 수정]
벡터 검색 (k=10) → RRF → TOP_K=3 결과 확정
→ TOP_K 결과 중 최고 score 기준으로 threshold 적용
→ 더 강한 후보만 살아남은 후 필터링 → 거절 기준 완화 효과
```

## Cell 5 — 검색 함수 정의

### 방법 D: TOP_K 결과 기준 threshold 필터 (IEP-2001.1 핵심 변경)

```
[IEP-2001 방법 C]                    [IEP-2001.1 방법 D]
질문                                  질문
 └─ 벡터 검색 k=10                    └─ 벡터 검색 k=10
     ↓ threshold 0.28 필터 (10개 전체)     ↓ (필터 없음)
     빈 리스트면 거절  ← 과다 거절       BM25 검색 k=10
     ↓ BM25 검색 k=10                     ↓ RRF 합산
     ↓ RRF 합산                           TOP_K=3 결과 확정
     TOP_K=3 반환                          ↓ 최고 score ≥ threshold?
                                          No → 거절  (더 관대한 기준)
                                          Yes → TOP_K 반환
```

**변경 포인트**: `hybrid_search()` 함수에서 threshold 필터 위치가
CANDIDATE_K 단계(10개) → TOP_K 단계(3개 중 최고값) 로 이동


````python
import os
from langchain_core.documents import Document

# ── 파라미터 ────────────────────────────────────────────────────────
TOP_K                = 3      # 최종 반환 청크 수
BM25_CANDIDATE_K     = 10     # BM25/벡터 각각에서 뽑는 후보 수
SIMILARITY_THRESHOLD = 0.28   # 벡터 score 임계값 (2026-04-22 확정)
RRF_K                = 60     # RRF 하이퍼파라미터 (표준값)
# ────────────────────────────────────────────────────────────────────


def vector_search(question: str, k: int = BM25_CANDIDATE_K):
    """벡터 검색 → List[(Document, cosine_score)]"""
    return vectorstore.similarity_search_with_relevance_scores(question, k=k)


def bm25_search(question: str, k: int = BM25_CANDIDATE_K):
    """BM25 검색 → List[(Document, bm25_score)]"""
    tokens = tokenize(question)
    scores = bm25.get_scores(tokens)
    top_indices = scores.argsort()[::-1][:k]
    results = []
    for idx in top_indices:
        doc = Document(
            page_content=docs_text[idx],
            metadata=docs_meta[idx]
        )
        results.append((doc, float(scores[idx])))
    return results


def rrf_fusion(vector_results, bm25_results, top_k: int = TOP_K):
    """
    Reciprocal Rank Fusion
    rrf_score = Σ 1 / (RRF_K + rank)  (rank: 0-indexed)
    두 리스트의 순위를 합산해 top_k개 반환
    """
    rrf_scores = {}   # doc_key → rrf_score 누적
    doc_map    = {}   # doc_key → Document
    vec_scores = {}   # doc_key → 원래 벡터 cosine score (threshold 판정용)

    def doc_key(doc: Document) -> str:
        return doc.page_content[:80]

    for rank, (doc, score) in enumerate(vector_results):
        k_ = doc_key(doc)
        rrf_scores[k_] = rrf_scores.get(k_, 0.0) + 1.0 / (RRF_K + rank + 1)
        doc_map[k_] = doc
        vec_scores[k_] = score   # 벡터 score 보존

    for rank, (doc, _) in enumerate(bm25_results):
        k_ = doc_key(doc)
        rrf_scores[k_] = rrf_scores.get(k_, 0.0) + 1.0 / (RRF_K + rank + 1)
        if k_ not in doc_map:
            doc_map[k_] = doc

    sorted_keys = sorted(rrf_scores, key=rrf_scores.get, reverse=True)
    top_keys = sorted_keys[:top_k]
    return [(doc_map[k_], rrf_scores[k_], vec_scores.get(k_, 0.0)) for k_ in top_keys]


def hybrid_search(question: str):
    """
    하이브리드 검색 메인 함수 (방법 D: TOP_K 결과 기준 threshold 필터)

    IEP-2001 방법 C 문제:
      CANDIDATE_K=10 전체에 threshold 적용 → borderline 질문 과다 거절 (17/100)

    IEP-2001.1 방법 D 변경:
      1. 벡터 + BM25 → RRF → TOP_K=3 결과 먼저 확정
      2. TOP_K 결과 중 최고 벡터 score >= SIMILARITY_THRESHOLD 이면 통과
      3. 최고 score도 미달이면 거절 (더 관대한 기준)

    반환: List[(Document, rrf_score)]  (rrf_score만 — 외부 인터페이스 동일)
    """
    # 1. 벡터 검색 (필터 없이 전체 후보 수집)
    v_results = vector_search(question, k=BM25_CANDIDATE_K)

    # 2. BM25 검색
    b_results = bm25_search(question, k=BM25_CANDIDATE_K)

    # 3. RRF 합산 → TOP_K 결과 확정 (vec_score 함께 반환)
    top_results = rrf_fusion(v_results, b_results, top_k=TOP_K)

    if not top_results:
        return []

    # 4. ★ 핵심 변경: TOP_K 결과 중 최고 벡터 score 기준으로 거절 판정
    best_vec_score = max(vec_s for _, _, vec_s in top_results)
    if best_vec_score < SIMILARITY_THRESHOLD:
        return []   # 범위 밖 질문 → 거절

    # 5. 외부 인터페이스 유지: (doc, rrf_score) 튜플로 반환
    return [(doc, rrf_s) for doc, rrf_s, _ in top_results]


print("검색 함수 정의 완료 (IEP-2001.1 방법 D)")
print(f"  TOP_K={TOP_K}, CANDIDATE_K={BM25_CANDIDATE_K}, THRESHOLD={SIMILARITY_THRESHOLD}, RRF_K={RRF_K}")
print(f"  거절 기준: TOP_K 결과 중 최고 벡터 score < {SIMILARITY_THRESHOLD}")
````

## Cell 6 — 단일 질문 검증 (방법 C vs 방법 D 비교)

**핵심 확인 포인트**:
- `환경 보호 개인 실천` (score 0.2624) — 방법 C: 거절 → 방법 D: **통과 여부 확인**
- borderline 질문들의 처리 차이 수치로 직접 확인
- 완전 범위 밖(`오늘 점심 메뉴`)은 방법 D에서도 여전히 거절되는지 확인


````python
test_questions = [
    ("1.5°C 목표를 달성하기 위해 필요한 탄소 감축량은?",  "수치/고유명사"),
    ("기후변화가 생태계에 미치는 영향은?",               "의미적 질문"),
    ("환경 보호를 위해 개인이 할 수 있는 일은?",         "경계 케이스 (score 0.2624)"),  # IEP-2001에서 거절됐던 질문
    ("해수면 상승이 연안 도시에 미치는 영향은?",         "일반 IPCC 질문"),
    ("오늘 점심 메뉴 추천해줘",                         "완전 범위 밖"),
]

print("방법 C (IEP-2001, CANDIDATE_K 기준) vs 방법 D (IEP-2001.1, TOP_K 기준) 비교")
print("=" * 75)

for q, label in test_questions:
    print(f"\n[{label}]")
    print(f"질문: {q}")

    # ── 방법 C 재현 (CANDIDATE_K 전체에 threshold 적용) ──────────
    v_all = vector_search(q, k=BM25_CANDIDATE_K)
    v_filtered_c = [(doc, s) for doc, s in v_all if s >= SIMILARITY_THRESHOLD]

    if not v_filtered_c:
        method_c = "거절"
        method_c_detail = f"최고 score={max(s for _,s in v_all):.4f} < {SIMILARITY_THRESHOLD}"
    else:
        b_res = bm25_search(q, k=BM25_CANDIDATE_K)
        rrf_c = rrf_fusion(v_filtered_c, b_res, top_k=TOP_K)
        method_c = f"통과 ({len(rrf_c)}개)"
        method_c_detail = f"통과 벡터 {len(v_filtered_c)}/10개"

    # ── 방법 D (현재 hybrid_search) ──────────────────────────────
    h_results = hybrid_search(q)

    if not h_results:
        # 거절된 경우 — 최고 score 표시를 위해 내부 계산
        v_all2 = vector_search(q, k=BM25_CANDIDATE_K)
        b_res2 = bm25_search(q, k=BM25_CANDIDATE_K)
        top3 = rrf_fusion(v_all2, b_res2, top_k=TOP_K)
        best_v = max((vs for _, _, vs in top3), default=0.0)
        method_d = "거절"
        method_d_detail = f"TOP_K 최고 벡터 score={best_v:.4f} < {SIMILARITY_THRESHOLD}"
    else:
        # 통과된 경우 — TOP_K 결과와 score 표시
        v_all2 = vector_search(q, k=BM25_CANDIDATE_K)
        b_res2 = bm25_search(q, k=BM25_CANDIDATE_K)
        top3_detail = rrf_fusion(v_all2, b_res2, top_k=TOP_K)
        best_v = max((vs for _, _, vs in top3_detail), default=0.0)
        method_d = f"통과 ({len(h_results)}개)"
        method_d_detail = f"TOP_K 최고 벡터 score={best_v:.4f}"

    # ── 결과 출력 ─────────────────────────────────────────────────
    verdict = ""
    if method_c == "거절" and method_d != "거절":
        verdict = "  ★ 방법 D에서 거절 → 통과 (개선)"
    elif method_c == "거절" and method_d == "거절":
        verdict = "  ✓ 둘 다 거절 (일치)"
    elif method_c != "거절" and method_d != "거절":
        verdict = "  ✓ 둘 다 통과 (일치)"
    else:
        verdict = "  ⚠️ 방법 D에서 통과 → 거절 (역전, 확인 필요)"

    print(f"  방법 C (IEP-2001) : {method_c:12}  [{method_c_detail}]")
    print(f"  방법 D (IEP-2001.1): {method_d:12}  [{method_d_detail}]")
    print(verdict)

    # ── 방법 D 통과 시 검색 결과 미리보기 ─────────────────────────
    if h_results:
        print(f"  방법 D 검색 결과:")
        for doc, rrf_s in h_results:
            page = doc.metadata.get('page', '?')
            print(f"    rrf={rrf_s:.4f} | page={page} | {doc.page_content[:65].strip()}...")

print("\n" + "=" * 75)
print("검증 완료 — 거절 기준 변화를 확인했으면 Cell 7 진행")
````

## Cell 7 — 골든 데이터셋 로드 + 100개 하이브리드 검색 실행

````python
import csv

with open(GOLDEN_PATH, "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    golden = list(reader)

print(f"골든 데이터셋: {len(golden)}개")
print(f"키 확인: {list(golden[0].keys())}")
print(f"샘플 question: {golden[0]['user_input'][:60]}")
print(f"샘플 reference: {golden[0]['reference'][:60]}")
````

````python
import time

print("100개 하이브리드 검색 실행 중...")
t0 = time.time()

hybrid_dataset = []
rejected_count = 0

for i, item in enumerate(golden):
    q            = item["user_input"]       # ← question → user_input
    ground_truth = item["reference"]        # ← ground_truth → reference

    results  = hybrid_search(q)
    contexts = [doc.page_content for doc, _ in results]

    if not contexts:
        rejected_count += 1

    hybrid_dataset.append({
        "question":        q,
        "ground_truth":    ground_truth,
        "contexts":        contexts,
        "retrieved_count": len(contexts),
        "rejected":        len(contexts) == 0,
    })

    if (i + 1) % 20 == 0:
        elapsed = time.time() - t0

elapsed = time.time() - t0
avg_chunks = sum(r["retrieved_count"] for r in hybrid_dataset) / len(hybrid_dataset)
print(f"거절 수: {rejected_count}/100")
print(f"평균 검색 청크 수: {avg_chunks:.2f}")
````

## Cell 8 — Solar 답변 생성 (100개)

기존 `IEP1001_case3_ragas_solar.ipynb`와 동일한 프롬프트 사용

````python
from openai import OpenAI
import time

SOLAR_BASE_URL = os.environ.get("SOLAR_BASE_URL", "https://api.upstage.ai/v1")
SOLAR_MODEL    = "solar-pro3"

solar_client = OpenAI(
    api_key=UPSTAGE_API_KEY,
    base_url=SOLAR_BASE_URL
)

RAG_PROMPT = """당신은 IPCC AR6 기후변화 보고서 전문 안내 AI입니다.
반드시 아래 제공된 문서 내용만을 근거로 답변하세요.
문서에 없는 내용은 추측하지 말고 "해당 내용은 IPCC 보고서에서 찾을 수 없습니다."라고 답하세요.

[참고 문서]
{context}

[질문]
{question}

[답변]"""

REJECTION_MESSAGE = (
    "죄송합니다. 해당 질문은 IPCC AR6 보고서의 범위를 벗어납니다. "
    "IPCC 기후변화 관련 질문을 입력해 주세요."
)

def generate_answer(question: str, contexts: list) -> str:
    if not contexts:
        return REJECTION_MESSAGE
    context_text = "\n\n".join(
        f"[청크 {i+1}]\n{c}" for i, c in enumerate(contexts)
    )
    prompt = RAG_PROMPT.format(context=context_text, question=question)
    resp = solar_client.chat.completions.create(
        model=SOLAR_MODEL,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=800
    )
    return resp.choices[0].message.content


print("Solar 답변 생성 중 (100개)...")
t0 = time.time()

for i, item in enumerate(hybrid_dataset):
    item["answer"] = generate_answer(item["question"], item["contexts"])
    if (i + 1) % 10 == 0:
        elapsed = time.time() - t0

total = time.time() - t0
print(f"\n답변 생성 완료: {total:.0f}초")
````
